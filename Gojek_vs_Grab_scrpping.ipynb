```python
from google_play_scraper import reviews, Sort
import pandas as pd

# Gojek
gojek_reviews, _ = reviews(
    'com.gojek.app',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=5000
)

# Grab
grab_reviews, _ = reviews(
    'com.grabtaxi.passenger',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=5000
)

gojek_df = pd.DataFrame(gojek_reviews)
grab_df = pd.DataFrame(grab_reviews)

gojek_df['platform'] = 'Gojek'
grab_df['platform'] = 'Grab'

df = pd.concat([gojek_df, grab_df], ignore_index=True)
```

```python
complaint_df = df[df['score'] <= 3]
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
df_clean = df[['platform', 'content', 'score', 'thumbsUpCount', 'at']]
```

```python
df_clean = df_clean.rename(columns={
    'content': 'review',
    'score': 'rating',
    'thumbsUpCount': 'engagement',
    'at': 'date'
})
```

```python
df_clean.to_csv('gojek_grab_reviews.csv', index=False, encoding='utf-8-sig')
```
